```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C)
plt.plot(X, S)

plt.show()
```

```python
"""
line_up, = plt.plot([1,2,3], label='Line 2')
line_down, = plt.plot([3,2,1], label='Line 1')
plt.legend(handles=[line_up, line_down])
"""

X = np.linspace(0.000000000001, 10, 100, endpoint=True)
logp, logn = np.log(X), -np.log(X)


p_log, = plt.plot(X, logp, label="p log")
n_log, = plt.plot(X, logn, label="n log")
plt.legend(handles=[p_log, n_log])
#legend((line1, line2, line3), ('label1', 'label2', 'label3'))
plt.show()
```

```python
import math
"""
line_up, = plt.plot([1,2,3], label='Line 2')
line_down, = plt.plot([3,2,1], label='Line 1')
plt.legend(handles=[line_up, line_down])
"""

X = np.linspace(0.000000000001, 100, 100, endpoint=True)
X = list(X)
logp = [math.log(i) for i in X]
loghalf = [math.log(i,1/2) for i in X ] 
loghn = [-math.log(i) for i in X ] 

p_log, = plt.plot(X, logp, label="p log")
half_log, = plt.plot(X, loghalf, label="half log")
n_log, = plt.plot(X, logn, label="n log")
plt.legend(handles=[p_log,half_log, n_log])
#legend((line1, line2, line3), ('label1', 'label2', 'label3'))
plt.show()
```
